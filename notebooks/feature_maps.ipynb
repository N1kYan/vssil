```python
# Imports etcs
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read feature map data
source_maps = torch.from_numpy(np.load("/home/yannik/vssil/source_map_k0.npy"))
print(source_maps.shape)
```

```python
plt.imshow(source_maps, cmap='gray')
plt.show()
```

```python
region_size = (32, 32)
h_start, w_start = 0, 0
fig, ax = plt.subplots(1, 1)
ax.imshow(source_maps, cmap='gray')
while True:
    max_loc = np.argmax(source_maps[h_start:h_start + region_size[0], w_start: w_start + region_size[1]])
    max_loc_h = (max_loc / source_maps.shape[0])
    max_loc_w = max_loc % source_maps.shape[1]
    print(f"h: {max_loc_h} | w: {max_loc_w}")
    ax.scatter(max_loc_h, max_loc_w, color='lime')
    h_start += region_size[0]
    if h_start >= source_maps.shape[0]:
        h_start = 0
        w_start += region_size[1]
        if w_start >= source_maps.shape[1]:
            break                                
```
